```
import pandas as pd
from sklearn.impute import SimpleImputer
import numpy as np
```

# 2. Feature Scaling

```
data = pd.read_csv("/content/drive/MyDrive/Data Mining 2021/1. Load data & Preprocessing/01. Data Preprocessing/Data.csv")
```

```
data.head()
```

```
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(data.iloc[:,1:3].values)
data.iloc[:,1:3]=imputer.transform(data.iloc[:,1:3].values)
data.head()
```

```
from sklearn.preprocessing import StandardScaler, Normalizer, MinMaxScaler

data = data.iloc[:,1:3]
standard_scaler = StandardScaler()
nomralizer = Normalizer()
min_max_scaler = MinMaxScaler()

#Standard scaler
print('Standarization')
print(standard_scaler.fit_transform(data))
```

```
#Normalizer
print('Normalizing')
print(nomralizer.fit_transform(data))
```

```
#Minmax scaling
print('Minmax Scaling')
print(min_max_scaler.fit_transform(data))
```
