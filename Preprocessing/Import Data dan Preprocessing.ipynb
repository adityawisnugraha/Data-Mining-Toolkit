```
import pandas as pd
from sklearn.impute import SimpleImputer
import numpy as np
```

# 1. Import Data dan Preprocessing

```
data = pd.read_csv("/content/drive/MyDrive/Data Mining/Data (1).csv")
```

```
data.head()
```

```
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(data.iloc[:,1:3].values)
data.iloc[:,1:3]=imputer.transform(data.iloc[:,1:3].values)
data.head()
```

## Encoding categorical data

```
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```
lable_encoder = LabelEncoder()
temp = data.copy()
temp.iloc[:, 0] = lable_encoder.fit_transform(data.iloc[:, 0])
temp.head()
```
